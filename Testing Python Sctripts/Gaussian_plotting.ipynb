```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
%matplotlib widget

#%matplotlib inline - cant rotate
#%matplotlib notebook

fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection ='3d')

x = list(np.random.randint(low=0,high=20,size=50))
y = list(np.random.randint(low=0,high=20,size=50))
z = list(np.random.randint(low=0,high=20,size=50))

ax.scatter3D(x,y,z, color='red', marker='o')

ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
ax.set_zlabel("z axis")
plt.title("3D plot")

plt.show()
```

```python
def plot_3d(x,y,z,color='red',marker='o'):

    """
    color = ['red', 'green', 'blue', 'purple']
    marker = ['o', 's', '^', 'p']
    """

    fig = plt.figure(figsize=(10,8))
    ax = plt.axes(projection ='3d')

    ax.scatter3D(x,y,z, color=color, marker=marker)

    ax.set_xlabel("x axis")
    ax.set_ylabel("y axis")
    ax.set_zlabel("z axis")
    plt.title("3D plot")

    ax.set_xlim3d(-4,4)
    ax.set_ylim3d(-4,4)
    ax.set_zlim3d(-4,4)

    plt.show()
    
```

```python
import numpy as np

m = np.array([0,0,0]).reshape(3,)
cov = np.array([[1,.5,0],
                [0.5,0.5,0],
                [0,0,1]])

z = np.random.multivariate_normal(mean=m, cov=cov, size = 1000)

print(z.shape)
```

```python
plot_3d(z[:,0],z[:,1],z[:,2])
```
